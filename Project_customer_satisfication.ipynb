# Choosing features 

I applied Logistic Regression to check the prediction accuracy of each feature 

If accuracy is lower than 60%, the feature would be excluded from the model testing

```python
# importing libraries
from sklearn.model_selection import train_test_split
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_selector
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
from sklearn.linear_model import LogisticRegression

# definding target vatiable and features
df_cat = pd.get_dummies(df.select_dtypes('object'), drop_first=True)
df_cat = pd.concat([df_cat, df.select_dtypes('number')],  axis=1)

# split the data 
X = df_cat.drop('satisfaction_satisfied', axis = 1)
y = pd.get_dummies(df_cat['satisfaction_satisfied'], drop_first=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)

# running Lgistic Regression
pipe = make_pipeline(StandardScaler(), LogisticRegression())
from sklearn.metrics import accuracy_score
from sklearn.metrics import precision_score
from sklearn.metrics import f1_score


# imputing missing data ('Arrival Delay in Minutes')
# 'Arrival Delay in Minutes' can be predicted by 'Departure Delay in Minutes'
# So 'Arrival Delay in Minutes' is imputed based on the value of 'Arrival Delay in Minutes'

import math
#X_train[['Arrival Delay in Minutes', 'Departure Delay in Minutes']].corr()
df_ = X_train[['Arrival Delay in Minutes', 'Departure Delay in Minutes']].copy()
df_.dropna(inplace = True)

arr = np.array(df_['Arrival Delay in Minutes']).reshape(-1, 1) 
dep = np.array(df_['Departure Delay in Minutes']).reshape(-1, 1)

from sklearn.linear_model import LinearRegression
reg = LinearRegression()
result = reg.fit(dep, arr)
coef = result.coef_
inter = result.intercept_

def cal_arr(dep):
    return dep*coef + inter


for i, j, ind in zip(X_train['Arrival Delay in Minutes'], X_train['Departure Delay in Minutes'], X_train.index):
    if math.isnan(i)== True:
        X_train.loc[ind, ['Arrival Delay in Minutes']] = cal_arr(j)[0][0]     
        
# Running LogisticRegression for each feature
# keeping the features its accuracy is more than 60%

acc_plus_60 = []

for i in X_train.columns:
    x = np.array(X_train[i]).reshape(-1,1)
    pipe.fit(x, y_train)
    #print(str(i), 'accuracy', accuracy_score(y_train, pipe.predict(x)))
    #print(str(i), 'precision', precision_score(y_train, pipe.predict(x)))
    #print(str(i), 'f1', f1_score(y_train, pipe.predict(x)))
    #print('######################################################################')
    if accuracy_score(y_train, pipe.predict(x)) > 0.6:
        acc_plus_60.append(i)
acc_plus_60
```

## Train test split 

```python
df = pd.read_csv('train.csv', index_col = 'id')
df.drop(columns = 'Unnamed: 0', inplace = True)

from sklearn.model_selection import train_test_split

X = df[['Type of Travel',
 'Class',
 'Flight Distance',
 'Inflight wifi service',
 'Ease of Online booking',
 'Online boarding',
 'Seat comfort',
 'Inflight entertainment',
 'On-board service',
 'Leg room service',
 'Baggage handling',
 'Checkin service',
 'Inflight service',
 'Cleanliness']]
y = pd.get_dummies(df['satisfaction'], drop_first=True)
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
```

## Preprocessing

```python
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_selector
from sklearn.compose import make_column_transformer
from sklearn.pipeline import make_pipeline
from sklearn.impute import SimpleImputer
```

```python
scaler = StandardScaler()
encoder = OneHotEncoder(drop= 'if_binary')

num_cols = make_column_selector(dtype_include='number')
cat_cols = make_column_selector(dtype_include='object')


num_transformer = make_pipeline(scaler)
cat_transformer = make_pipeline(encoder)


preprocessor = make_column_transformer((num_transformer, num_cols),
                                      (cat_transformer, cat_cols))
```

## Model Selection

```python
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import GridSearchCV
```

```python
models = [KNeighborsClassifier(),
          GaussianNB()]
for model in models:
    clf = model
    pipe = make_pipeline(preprocessor, clf)
    grip_a = GridSearchCV(estimator=pipe, param_grid={}, cv = 5, scoring = 'accuracy')
    grip_a.fit(X_train, y_train)
    
    grip_p = GridSearchCV(estimator=pipe, param_grid={}, cv = 5, scoring = 'precision')
    grip_p.fit(X_train, y_train)
    
    print(model, 'accuracy', grip_a.best_score_)
    print(model, 'precision', grip_p.best_score_)
```

```python
models = [LogisticRegression(),
          DecisionTreeClassifier(),
          RandomForestClassifier()]
for model in models:
    clf = model
    pipe = make_pipeline(preprocessor, clf)
    grip_a = GridSearchCV(estimator=pipe, param_grid={}, cv = 5, scoring = 'accuracy')
    grip_a.fit(X_train, y_train)
    
    grip_p = GridSearchCV(estimator=pipe, param_grid={}, cv = 5, scoring = 'precision')
    grip_p.fit(X_train, y_train)
    
    print(model, 'accuracy', grip_a.best_score_)
    print(model, 'precision', grip_p.best_score_)
```

```python
models = [SVC()]

for model in models:
    clf = model
    pipe = make_pipeline(preprocessor, clf)
    grip_a = GridSearchCV(estimator=pipe, param_grid={}, cv = 5, scoring = 'accuracy')
    grip_a.fit(X_train, y_train)
    
    grip_p = GridSearchCV(estimator=pipe, param_grid={}, cv = 5, scoring = 'precision')
    grip_p.fit(X_train, y_train)
    
    print(model, 'accuracy', grip_a.best_score_)
    print(model, 'precision', grip_p.best_score_)
    
```
